# QuantStats Tutorial

```python
import quantstats as qs
```

```python
# fetch the daily returns for a stock and a benchmark
stock = qs.utils.download_returns('META')
bench = qs.utils.download_returns('SPY')

# known bug requires timezone naive index
stock.index = stock.index.tz_convert(None)
bench.index = bench.index.tz_convert(None)
```

```python
# view all the metrics
[f for f in dir(qs.stats) if f[0] != '_']
```

```python
# show sharpe ratio
qs.stats.sharpe(stock)
```

```python
# view all the plots
[f for f in dir(qs.plots) if f[0] != '_']
```

```python
qs.plots.rolling_volatility(stock)
```

```python
# create a quick snapshot
qs.plots.snapshot(stock, title='Facebook Performance')
```

```python
# print out a metrics report - works with strategy returns too!
qs.reports.metrics(
    stock,
    bench,
    mode="full"
)
```

```python
# print out a full report
# to export to HTML:
#   qs.reports.html(stock, bench)
qs.reports.full(stock, bench)
```
